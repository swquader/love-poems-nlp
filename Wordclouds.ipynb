```python
import pandas as pd
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle('/Users/wasilaq/Metis/love-poems-nlp/pickled/final_poems_df')

post2000_words = ' '.join(
    [x for x in df['poem_clean'] if df.loc[df['poem_clean']==x]['period'].values[0]=='post-2000']
)

pre2000_words = ' '.join(
    [x for x in df['poem_clean'] if df.loc[df['poem_clean']==x]['period'].values[0]=='pre-2000']
)

# add stop words that were added in topic modeling
sw = list(STOPWORDS) + ['love','like','know','shall','heart','eyes','loves','light','thou','thee','thy','yo']
```

```python
pre2000_words = pre2000_words.replace('love','')
```

```python
wordcloud_pre2000 = WordCloud(
    width = 800, height = 800, background_color ='white', stopwords = set(sw), min_font_size = 10
).generate(pre2000_words) 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud_pre2000) 
plt.axis("off") 
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
wordcloud_post2000 = WordCloud(
    width = 800, height = 800, background_color ='white', stopwords = set(sw), min_font_size = 10
).generate(post2000_words) 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud_post2000) 
plt.axis("off") 
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
wordcloud = WordCloud(
    width = 800, height = 800, background_color ='white', stopwords = set(sw), min_font_size = 10
).generate(pre2000_words+post2000_words) 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)
  
plt.show()
```

```python
# only use standard stop words

wordcloud = WordCloud(
    width = 800, height = 800, background_color ='white', stopwords = set(STOPWORDS), min_font_size = 10
).generate(pre2000_words+post2000_words) 

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0)
  
plt.show()
```
